```python


# Get the movies rated by user 1
user1_movies = moviemat.loc[1].dropna().index

# Initialize a dictionary to store the count of common movies for each user
common_movies_count = {}

# Loop through all users
for user_id, ratings in moviemat.iterrows():
    if user_id != 1:
        # Get the movies rated by the current user
        user_movies = ratings.dropna().index
        
        # Count the number of common movies with user 1
        common_count = sum(movie in user_movies for movie in user1_movies)
        
        # Store the count in the dictionary
        common_movies_count[user_id] = common_count

# Find the user with the highest number of common movies
most_common_user = max(common_movies_count, key=common_movies_count.get)
num_common_movies = common_movies_count[most_common_user]

print(f"User {most_common_user} has the highest number of common movies ({num_common_movies}) rated with User 1.")
print(f"Total number of common movies: {num_common_movies}")
```

```python
# Get the movies rated by user 1 and user 414
user1_movies = set(moviemat.loc[1].dropna().index)
user414_movies = set(moviemat.loc[414].dropna().index)

# Find the common movies between user 1 and user 414
common_movies_user1_414 = user1_movies.intersection(user414_movies)

# Initialize a dictionary to store the count of common movies for each user
common_movies_count = {}

# Loop through all users
for user_id, ratings in moviemat.iterrows():
    if user_id != 1 and user_id != 414:
        # Get the movies rated by the current user
        user_movies = set(ratings.dropna().index)
        
        # Count the number of common movies with user 1 and user 414
        common_count = len(common_movies_user1_414.intersection(user_movies))
        
        # Store the count in the dictionary
        common_movies_count[user_id] = common_count

# Find the user with the highest number of common movies
third_common_user = max(common_movies_count, key=common_movies_count.get)
num_common_movies_third_user = common_movies_count[third_common_user]



print(f"\nUser {third_common_user} has the highest number of common movies ({num_common_movies_third_user}) with the common movies between Users 1 and 414.")
```

```python
import pandas as pd

# Assuming moviemat is your pivot table
# moviemat = data.pivot_table(index='userId', columns='title', values='rating')

# Get the movies rated by user 1, user 414, and user 599
user1_movies = set(moviemat.loc[1].dropna().index)
user414_movies = set(moviemat.loc[414].dropna().index)
user599_movies = set(moviemat.loc[599].dropna().index)

# Find the common movies between all three users
common_movies_all_users = user1_movies.intersection(user414_movies, user599_movies)

# Extract the ratings for common movies and store in a table (DataFrame)
common_movies_table = moviemat.loc[[1, 414, 599], list(common_movies_all_users)].fillna(np.nan)

print("Table of common movies and their ratings for all three users:")
common_movies_table
```

```python
# Create a new DataFrame for Borda count scores
borda_count_table = pd.DataFrame(index=common_movies_table.index, columns=common_movies_table.columns)

# Total number of movies
total_movies = len(borda_count_table.columns)

# Iterate through columns
for movie in borda_count_table.columns:
    # Get the indices of non-NaN values sorted in descending order
    sorted_indices = np.argsort(common_movies_table[movie].dropna().values)[::-1]
    
    # Calculate Borda count scores
    borda_count_scores = np.arange(total_movies, 0, -1)
    
    # Create a dictionary for mapping indices to Borda count scores
    score_dict = dict(zip(common_movies_table.index[sorted_indices], borda_count_scores))
    
    # Assign Borda count scores to the new DataFrame using .loc
    borda_count_table.loc[:, movie] = borda_count_table.index.to_series().map(score_dict)

borda_count_table = borda_count_table.astype(float)  # Ensure the dtype is float
borda_count_table
```

```python
# Calculate the mean rating for each movie
mean_ratings = borda_count_table.mean(axis=0)

# Create a dictionary to store the average value and the title of the movie
average_ratings_dict = {'Title': mean_ratings.index, 'Average Rating': mean_ratings.values}

# Sort the movies by mean rating in descending order
sorted_mean_ratings = mean_ratings.sort_values(ascending=False)

# Create a dictionary to store the top 10 highest-rated movies
top_10_movies_dict = {'Title': sorted_mean_ratings.head(10).index, 'Average Rating': sorted_mean_ratings.head(10).values}
```

```python
# Assuming borda_count_table is your DataFrame

# Calculate the minimum rating given by each user for each movie
min_ratings_by_user = borda_count_table.min(axis=0)

# Sort the movies by the minimum rating in descending order (highest ratings among the lowest)
sorted_min_ratings_least = min_ratings_by_user.sort_values(ascending=False)

# Create a dictionary to store all movies
all_least_misery_dict = {'Title': sorted_min_ratings_least.index.tolist(),
                         'Least Misery Rating': sorted_min_ratings_least.values.tolist()}


all_least_misery_dict
```
